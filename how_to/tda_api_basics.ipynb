# Price History Endpoint

[tda api doc for price history](https://developer.tdameritrade.com/price-history/apis/get/marketdata/%7Bsymbol%7D/pricehistory)

```python
symbol = 'AAPL'
endpoint = f"https://api.tdameritrade.com/v1/marketdata/{symbol}/pricehistory"

payload = {'apikey': client_id,
           'periodType': 'year',
           'frequencyType': 'daily',
           'frequency': '1',
           'period': '2',
           'needExtendedHoursData': 'true'
          } 

content = requests.get(url=endpoint, params = payload)
data = content.json()
data
```

# Quotes

```python
endpoint = 'https://api.tdameritrade.com/v1/marketdata/quotes'

payload = {'apikey': client_id,
           'symbol': 'AAPL,GOOG,FB,TSLA,NFLX,RBLX'}

content = requests.get(url=endpoint, params = payload)
data = content.json()
data
```
